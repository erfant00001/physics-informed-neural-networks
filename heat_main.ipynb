```python
import numpy 
from matplotlib import pyplot
```

```python
length = 10
k = 0.89
# B.C
temp_left = 100
temp_right = 200
total_sim_time = 10
```

```python
dx = .1
x_vector = numpy.linspace(0, length, int(length/dx))
```

```python
x_vector
```

```python
dt = 0.0001
t_vector = numpy.linspace(0, total_sim_time, int(total_sim_time/dt))
```

```python
t_vector
```

```python
# I.C
u = numpy.zeros([len(t_vector),len(x_vector)])
```

```python
u.shape
```

```python
u
```

```python
u[0,0] =100
```

```python
u
```

```python
u[:,0] = temp_left
u[:,-1] = temp_right
```

```python
u
```

```python
pyplot.plot(x_vector,u[0])
pyplot.ylabel("temp")
pyplot.xlabel("x")
pyplot.show()
```

```python
for t in range(1, len(t_vector)-1):
    for x in range(1, len(x_vector)-1):
        u[t+1,x] = ((k* (dt/dx**2)) * (u[t, x+1] - 2* u[t,x] + u[t,x-1])) + u[t,x]
```

```python
u
```

```python
pyplot.plot(x_vector,u[len(u)-1, :])
pyplot.ylabel("temp")
pyplot.xlabel("x")
pyplot.show()
```

```python
len(u)
```

```python
for i in range(0, len(u), 5000):
    pyplot.plot(x_vector, u[i])

pyplot.xlabel("x")
pyplot.ylabel("Temperature")
pyplot.title("Temperature vs. x at different time steps")
pyplot.grid(True)
pyplot.show()
```
